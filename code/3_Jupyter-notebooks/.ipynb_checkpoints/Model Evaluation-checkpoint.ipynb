## Random Survival Forest Model Evaluation

Documentation for evalutaion metrics: https://scikit-survival.readthedocs.io/en/latest/api/metrics.html

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import eli5
from eli5.sklearn import PermutationImportance
from sklearn.model_selection import train_test_split
import sksurv.datasets
from sksurv.ensemble import RandomSurvivalForest
from sksurv.metrics import brier_score, concordance_index_censored, concordance_index_ipcw, cumulative_dynamic_auc, integrated_brier_score


%matplotlib inline
```

```python
model = joblib.load('../Google_Cloud_Platform/trained_model/model.joblib')
```

```python
df = pd.read_csv('../Google_Cloud_Platform/trained_model/df_gcp.csv')
df.set_index(keys = 'TRR_ID_CODE', inplace = True)


X, y = sksurv.datasets.get_x_y(df, attr_labels = ['GRF_STAT_NUM', 'GTIME_KI_YEARS'], 
                              pos_label = 1, survival = True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.33, random_state = 39)
```

```python
df.shape
```

```python
survs = model.predict_survival_function(X_test)
```

```python
# Plot predicted survival functions for 5 patients

surv_funcs = model.predict_survival_function(X_test.iloc[:5], return_array = True)

for i, s in enumerate(surv_funcs):
    plt.step(model.event_times_, s, where = 'post', label = str(i))
plt.ylabel('Survival Probability')
plt.xlabel("Time in Years")
plt.legend()
plt.grid(True)
```

```python
# define parameters used in final model
model.__init__
```

```python
# plot predicted cumulative hazard function for 10 patients:
cum_surv = model.predict_cumulative_hazard_function(X_test.iloc[:10], return_array = True)

for i, s in enumerate(cum_surv):
    plt.step(model.event_times_, s, where = 'post', label = str(i))
plt.ylabel("Cumulative Hazard")
plt.xlabel("Time in Years")
plt.grid(True)
```

### Concordance Index

```python
# Concordance index of prediction (Harrell's)
model.score(X_test, y_test)
```

```python
# Concordance index ipcw:  does not depend on distribution of censoring times in the test data. 
#Estimate is unbiased and consistent for a population concordance measure that is free of censoring

# 5 years
preds = [fn[5] for fn in survs]

concordance_index_ipcw(y_train, y_test, preds, 5)
```

```python
# 10 years
preds = [fn[10] for fn in survs]

concordance_index_ipcw(y_train, y_test, preds, 10)
```

### Model Brier Score & Integrated Brier Score

Info on interpreting Brier Score: https://k-d-w.org/blog/2020/06/scikit-survival-0.13-released/

```python
# Brier score at 5 years

#survs = model.predict_survival_function(X_test)
preds = [fn[5] for fn in survs]

times, score = brier_score(y_train, y_test, preds, 5)
print(score)
```

```python
# Brier score at 10 years

preds = [fn[10] for fn in survs]

times, score = brier_score(y_train, y_test, preds, 10)
print(score)
```

```python
# Integrated Brier score (provides overall calculation of model performance at all available times)
# between 0 and 5 years after transplant
times = np.arange(0, 5)
preds = np.asarray([[fn[t] for t in times] for fn in survs])

integrated_brier_score(y_train, y_test, preds, times)
```

```python
# Integrated Brier score (provides overall calculation of model performance at all available times)
# between 0 and 10 years after transplant
times = np.arange(0, 10)
preds = np.asarray([[fn[t] for t in times] for fn in survs])

integrated_brier_score(y_train, y_test, preds, times)
```

### Cumulative Dynamic AUC

```python
times = np.arange(0,14,1) # assess how well the model distinguishes graft failure from survival in yearly intervals up to 14 years post-transplant

auc, mean_auc = cumulative_dynamic_auc(y_train, y_test, model.predict(X_test), times)

plt.plot(times, auc, marker="o")
plt.axhline(mean_auc, linestyle="--")
plt.xlabel("Post-transplant Years")
plt.ylabel("time-dependent AUC")

plt.grid(True)
```

```python
times = np.arange(0,5, 0.5) # assess how well the model distinguishes graft failure from survival in 6 month intervals up to 5 years post-transplant

auc, mean_auc = cumulative_dynamic_auc(y_train, y_test, model.predict(X_test), times)

plt.plot(times, auc, marker="o")
plt.axhline(mean_auc, linestyle="--")
plt.xlabel("Post-transplant Years")
plt.ylabel("time-dependent AUC")

plt.grid(True)
```

```python
times = np.arange(0,14,1) # assess how well the model distinguishes graft failure from survival in yearly intervals up to 14 years post-transplant

auc, mean_auc = cumulative_dynamic_auc(y_train, y_test, model.predict(X_test), times)

plt.plot(times, auc, marker="o")
plt.axhline(mean_auc, linestyle="--")
plt.xlabel("Post-transplant Years")
plt.ylabel("time-dependent AUC")

axes = plt.gca()
axes.set_ylim([0, 1.0])
axes.set_xlim([0,14])

plt.grid(True)
```

```python
times = np.arange(0,5, 1) # assess how well the model distinguishes graft failure from survival in yearly intervals up to 5 years post-transplant

auc, mean_auc = cumulative_dynamic_auc(y_train, y_test, model.predict(X_test), times)

plt.plot(times, auc, marker="o")
plt.axhline(mean_auc, linestyle="--")
plt.xlabel("Post-transplant Years")
plt.ylabel("time-dependent AUC")

axes = plt.gca()
axes.set_ylim([0, 1.0])
axes.set_xlim([0,5])

plt.grid(True)
```

```python
times = np.arange(0,5,0.5) # assess how well the model distinguishes graft failure from survival in 6 month intervals up to 5 years post-transplant

auc, mean_auc = cumulative_dynamic_auc(y_train, y_test, model.predict(X_test), times)

plt.plot(times, auc, marker="o")
plt.axhline(mean_auc, linestyle="--")
plt.xlabel("Post-transplant Years")
plt.ylabel("time-dependent AUC")

axes = plt.gca()
axes.set_ylim([0, 1.0])
axes.set_xlim([0,5])

plt.grid(True)
```

### Permutation Importance

```python
df = pd.read_csv('../Google_Cloud_Platform/trained_model/df_gcp.csv')
df.set_index(keys = 'TRR_ID_CODE', inplace = True)
```

```python

X, y = sksurv.datasets.get_x_y(df, attr_labels = ['GRF_STAT_NUM', 'GTIME_KI_YEARS'], 
                              pos_label = 1, survival = True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.1, random_state = 39)
```

```python
#https://medium.com/towards-artificial-intelligence/how-to-use-scikit-learn-eli5-library-to-compute-permutation-importance-9af131ece387

perm = PermutationImportance(model, random_state=93, n_iter = 5, cv = 3)
perm.fit(X_test, y_test)
```

```python
eli5.show_weights(perm, feature_names=list(X_test.columns), top = 100)
```

```python
weights = eli5.explain_weights_df(perm, top = 100,  feature_names = list(X_test.columns))
weights
```

```python
weights.to_csv('../../data/weights_10.csv')
```
