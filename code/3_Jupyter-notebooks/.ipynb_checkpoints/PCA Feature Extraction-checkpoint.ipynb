### Principal Component Analysis

Due to the high number of features, I will first use FAMD to extract a subset of features that predict graft failure for my model.

```python
# Create subest of dataset with only features to be used in model
# Removed KDRI/KDPI, all variables associated with graft failure, and ID's (tranplant, patient, etc.)
df_PCA = df.drop(columns = {'PX_STAT_DATE','MAX_KDPI_LOCAL_ZERO_ABDR',
'MAX_KDPI_LOCAL_NON_ZERO_ABDR','MAX_KDPI_IMPORT_NON_ZERO_ABDR','MAX_KDPI_IMPORT_ZERO_ABDR',
'TX_DATE','KDPI','KDRI_RAO','EGFR_CKDEPI_DON','KDRI_MED','GRF_STAT_KI','GSTATUS_KI',
'PSTATUS','PTIME','GTIME_KI','DWFG_KI',
'PX_STAT_DATE','INIT_DATE','WT_QUAL_DATE','TX_DATE','ADMISSION_DATE',
'RECOVERY_DATE','DON_DATE','BRAIN_DEATH_DATE', 'GRF_FAIL_DATE', 'FAILDATE_KI'})
```

```python
df_PCA.isnull().sum().sort_values(ascending = False)
```

```python
df_PCA.drop(columns = {'INIT_CPRA', 'OTHER_BLOOD', 'PT_T4', 'CARDARREST_DOWNTM', 'END_CPRA_DETAIL',
                      'END_CPRA', 'BRAIN_DEATH_TIME', 'CPR_ADMIN', 'WORK_INCOME_TCR',
                      'CHEST_TRAUMA','GSTATUS_DTHCNS_KI'},inplace = True)
```

```python
df_PCA.isnull().sum().sort_values(ascending = False)
```

I decided to drop null values for PCA, and then impute nulls after feature extraction, if needed. 

```python
df_PCA = df_PCA.dropna()
```

```python
df_PCA.shape
```

```python
y = df_PCA['GRF_STAT']
X = df_PCA.drop(columns = ['GRF_STAT'])

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.2,
                                                    random_state = 39)
```

```python
y_train.value_counts(normalize = True)
```

```python
y_test.value_counts(normalize = True)
```

```python
# Random Undersample X_train, y_train for class imbalance
nm = RandomUnderSampler()

X_train_under, y_train_under = nm.fit_resample(X_train, y_train)
```

```python
y_train_under.value_counts()
```

```python
y_test.value_counts(normalize = True)
```

```python
mm = MinMaxScaler()
X_train = mm.fit_transform(X_train_under)
X_test = mm.transform(X_test)
```

```python
# Baseline Training and Test Scores
lr = LogisticRegression(max_iter = 1000)
lr.fit(X_train, y_train_under)
lr.score(X_train, y_train_under), lr.score(X_test, y_test)
```

```python
pca = PCA(n_components = 500, random_state = 39)
pca.fit(X_train)

Z_train = pca.transform(X_train)
Z_test = pca.transform(X_test)
```

```python
# PCA Training and test scores
lr.fit(Z_train, y_train_under)
lr.score(Z_train, y_train_under), lr.score(Z_test, y_test)
```

```python
# Pull the explained variance attribute.
var_exp = pca.explained_variance_ratio_
print(f'Explained variance (first 500 components): {np.round(var_exp[:500],3)}')

print('')

# Generate the cumulative explained variance.
cum_var_exp = np.cumsum(var_exp)
print(f'Cumulative explained variance (first 500 components): {np.round(cum_var_exp[:500],3)}')
```
